```
# Assuming you have run the installation and path definition from Step 1

import os
import shutil

# ----------------- PATH DEFINITION (Recap) -----------------
ZIP_FILE_NAME = 'Combined_dataset.zip'
INITIAL_EXTRACT_PATH = '/content/Combined_dataset_TEMP'
DATASET_ROOT = '/content/Combined_dataset'

# ----------------- Step 2: Unzip and Fix Nested Structure ğŸ› ï¸ -----------------

# 1. Unzip the file to a temporary location
os.makedirs(INITIAL_EXTRACT_PATH, exist_ok=True)
!unzip -q {ZIP_FILE_NAME} -d {INITIAL_EXTRACT_PATH}

print("--- Contents of the temporary folder (INITIAL_EXTRACT_PATH) ---")
!ls {INITIAL_EXTRACT_PATH}

# 2. Dynamically find the name of the single nested folder
# This is usually the folder that holds train/test/valid
try:
    # List contents of the temp folder and assume the first folder listed is the nested one
    nested_contents = os.listdir(INITIAL_EXTRACT_PATH)

    # Filter out common non-data files/folders (like __MACOSX if zipped on a Mac)
    nested_folders = [d for d in nested_contents if os.path.isdir(os.path.join(INITIAL_EXTRACT_PATH, d)) and d != '__MACOSX']

    if not nested_folders:
        # If no nested folder is found, the contents are already at the top level
        print("\nStructure: Contents are already at the top level of the zip.")
        NESTED_DIR = INITIAL_EXTRACT_PATH
    elif len(nested_folders) == 1:
        # One nested folder found
        NESTED_DIR = os.path.join(INITIAL_EXTRACT_PATH, nested_folders[0])
        print(f"\nStructure: Nested folder found: {nested_folders[0]}")
    else:
        # Multiple folders found, handle carefully (assuming the largest one holds the data)
        # For simplicity, we'll assume the desired contents are directly in INITIAL_EXTRACT_PATH
        print("\nWarning: Multiple folders detected. Assuming contents are at top-level.")
        NESTED_DIR = INITIAL_EXTRACT_PATH

except Exception as e:
    print(f"\nError during folder detection: {e}. Assuming contents are at top-level.")
    NESTED_DIR = INITIAL_EXTRACT_PATH


# 3. Move all actual dataset files (train, test, valid, data.yaml) to the desired root
os.makedirs(DATASET_ROOT, exist_ok=True)

# Use rsync for a more robust move operation, or mv if rsync is unavailable/complex
# Using mv for simplicity, but pointing to the contents of the detected directory
!mv {NESTED_DIR}/* {DATASET_ROOT}/

# 4. Clean up the temporary folder
shutil.rmtree(INITIAL_EXTRACT_PATH, ignore_errors=True)

print(f"\nStructure fixed! Your dataset is now correctly at: {DATASET_ROOT}")

# 5. Verify the final contents
print("--- Final Cleaned Folder Contents ---")
!ls {DATASET_ROOT}
```
